```python
## RAG Demo: Retrieve Answers from PDF using Hugging Face

import os
from langchain_community.document_loaders import PyPDFLoader
from langchain.vectorstores import FAISS
from langchain.chains import RetrievalQA
from langchain.embeddings import HuggingFaceEmbeddings
from transformers import pipeline

# Load the PDF document
pdf_path = 'Curated Learning Resources for Key Roles.pdf'
loader = PyPDFLoader(pdf_path)
documents = loader.load()

# Split and embed the document
embedding_model = HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')
vector_store = FAISS.from_documents(documents, embedding_model)

# Set up the Hugging Face QA model
qa_model = pipeline('text2text-generation', model='google/flan-t5-base')

# RAG function to retrieve context and generate answers
def ask_question(question):
    docs = vector_store.similarity_search(question, k=3)
    context = '\n'.join([doc.page_content for doc in docs])

    prompt = f"Context: {context}\n\nQuestion: {question}\nAnswer:"
    response = qa_model(prompt, max_length=256, truncation=True)

    return response[0]['generated_text']

# Example query
question = "What resources can I use to transition to a Software Engineer role?"
print("Q:", question)
print("A:", ask_question(question))
```
